```python
import cv2
import matplotlib.pyplot as plt
import os

root_path = "maize_dataset_split/train"
class_labels = sorted(os.listdir(root_path))
labels_mapping = {class_label:i for i,class_label in enumerate(class_labels)}
print(labels_mapping)
```

```python
# Checking the number of files in each class to check class imbalance
total_images = 0
for class_label in class_labels:
    number_of_files = len(os.listdir(f"{root_path}/{class_label}"))
    total_images += number_of_files
    print(f"Number of files : {number_of_files} | Label : {class_label}")
print(f"Total Images : {total_images}")
```

```python
# Checking the size of images in the dataset for processing
import os
import cv2

for class_label in class_labels:
    size_dict = {}
    size_set = set()
    folder_path = os.path.join(root_path, class_label)
    img_paths = os.listdir(folder_path)
    for img_name in img_paths:
        img_path = os.path.join(folder_path, img_name)
        img = cv2.imread(img_path)        
        size = img.shape  
        size_set.add(size)
        if size in size_dict:
            size_dict[size] += 1
        else:
            size_dict[size] = 1
    print("*"*50)
    print(f"Class label : {class_label}")
    print(size_dict)
```
